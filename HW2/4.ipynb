```python
from sklearn import datasets
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
data = iris.data[:,0]
historgrams = plt.hist(data,bins=100)
#plt.savefig('4-1')
plt.show()
```

```python
p = [  ]
for i in range(100):
    h =historgrams[1][i+1]  - historgrams[1][i] 
    y = historgrams[0][i] / (100* h )
    p.append(y)
plt.plot(p)
plt.ylim(0.1,3)
#plt.savefig('4-2')
plt.show()
```
