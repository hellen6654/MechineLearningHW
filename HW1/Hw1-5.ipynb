```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
array = np.random.rand(201)
result = []
for i in range(201):
    temp = array[i]**3 * (1-array[i])**2
    result.append(temp)
index = np.argmax(result)
array[index]
```

yes, 約莫等於0.6

```python
plt.grid()
plt.plot(array,result,'o') #畫全部的點
plt.plot(array[index],result[index],'ro') #畫最大值的那個點
plt.plot([array[index],array[index]],[0,result[index]],'r-') #畫最大值的那個點
plt.show()
```
